```python
import os
import networkx as nx
import math
import numpy as np
import pickle
from random import sample,choices,uniform, randint
from DSSCircuit_Interface import *
from Outage_Eval import *
```

```python
#Read the .gml graph and load it into notebook, get the node list
FolderName=os.path.dirname(os.path.realpath("__file__"))
G_original = nx.read_gml(""+ FolderName+ r"\8500busEx.gml")
nodes=list(G_original.nodes())
edges=list(G_original.edges())
#source_node = 'sourcebus'
```

```python
sensor_nodes=[] 
Nonsensor_nodes=[] 
sample_Nindx = sample(range(1,len(nodes)), int(0.5*len(nodes))) 
for s in sample_Nindx: 
    Nonsensor_nodes.append(nodes[s])
    
    filename = "./BuswoSensors.npy" 
    np.save(filename, Nonsensor_nodes)
```

```python
Nonsensor_edges = []
sample_EDindx = sample(range(1,len(edges)), int(0.5*len(edges))) 
for s in sample_EDindx: 
    name=G_original.edges[edges[s]]['device'] + '.' + G_original.edges[edges[s]]['label']
    Nonsensor_edges.append(name)
    filename = "./LinewoSensors.npy" 
    np.save(filename, Nonsensor_edges)
```
